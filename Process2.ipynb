```python
# Hapus data duplikat berdasarkan kolom "nama"
df_test = df_test.drop_duplicates(subset = "Model Name")

#Cek data duplikat lagi
print(df_test.duplicated().sum())
```

```python
# Ganti nama fitur "Manufactur" -> "Company"
df_test = df_test.rename(columns={'Manufacturer' : 'Company'})

# Ganti nama fitur "Category" -> "TypeName"
df_test = df_test.rename(columns={'Category' : 'TypeName'})

# Ganti nama fitur CPU" -> "Cpu"
df_test = df_test.rename(columns={"CPU": "Cpu"})

# Ganti nama fitur "RAM" -> "Ram"
df_test = df_test.rename(columns={"RAM": "Ram"})

# Ganti nama fitur "Storage" -> "Memory"
df_test = df_test.rename(columns={"Storage": "Memory"})

# Ganti nama fitur "GPU" -> "Gpu "
df_test = df_test.rename(columns={"GPU": "Gpu "})
```

```python
#Price Process
df_test['Price'] = df_test['Price'] / 100 

#INR to IDR
df_test['Price'] = df_test['Price'] * 188.31 
df_train['Price'] = df_train['Price'] * 188.31 

#Redominasi
df_test['Price'] = df_test['Price'] / 1000000
df_train['Price'] = df_train['Price'] / 1000000

#Deskripsi nilai Price
print("Train Price : ", df_train['Price'].describe())
print("Test Price : ", df_test['Price'].describe())
```
